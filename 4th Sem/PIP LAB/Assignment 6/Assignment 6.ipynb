##### 3. Create a dictionary whose keys are month names and whose values are the number of days in the corresponding months. 
(a) Ask the user to enter a month name and use the dictionary to tell how many days are in the month. 
(b) Print out all of the keys in alphabetical order. 
(c) Print out all of the months with 31 days. 
(d) Print out the (key-value) pairs sorted by the number of days in each month.

```python
months = {'January':31, 'February':28, 'March':31, 'April':30, 'May':31, 'June':30, 'July':31, 'August':31, 'September':30, 'October':31, 'November':30, 'December':31}
#a)
a = input("Enter a Month Name: ")
print("A) Days: ",months.get(a, "Invalid"))
#b)
print("B) Months in Alphabetical Order: ")
for k in sorted(months.keys()):
    print(k,end=" ")
#c)
print()
print("C) Months with 31 days: ")
for k, v in months.items():
    if v == 31:
        print(k,end=" ")
#d)
print()
print("D) Months Sorted by Number of Days: ")
for k, v in sorted(months.items(), key=lambda x:x[1]):
    print(k, v)
```

##### 6. Write a program that checks if two same values in a dictionary have different keys. That is, for dictionary D1 = { 'a' : 10, 'b': 20, 'c' : 10}, the program should print 2 keys have same values and for dictionary D2 = {'a' : 10, 'b' : 20, 'c' : 30} , the program should print No keys have same values. 

```python
D = {'a' : 10, 'b': 20, 'c' : 10}
#D = {'a' : 10, 'b' : 20, 'c' : 30}
values = list(D.values())
if len(values) != len(set(values)):
    print("No Keys Have Same Values.")
else:
    print("Keys Have Same Values.")
```

##### 10. Write a program to input your friends’ names and their Phone Numbers and store them in the dictionary as the key-value pair. Perform the following operations on the dictionary: a) Display the name and phone number of all your friends b) Add a new key-value pair in this dictionary and display the modified dictionary c) Delete a particular friend from the dictionary d) Modify the phone number of an existing friend e) Check if a friend is present in the dictionary or not f) Display the dictionary in sorted order of names

```python
friends = {}
n = int(input("Enter number of friends: "))
for i in range(n):
    name = input("Enter name: ")
    phone = input("Enter phone number: ")
    friends[name] = phone
print("\nFriends List:")
for name, phone in friends.items():
    print(name, ":", phone)
name = input("\nEnter new friend name: ")
phone = input("Enter phone number: ")
friends[name] = phone
print("Updated dictionary:", friends)
name = input("\nEnter friend name to delete: ")
friends.pop(name, None)
print("After deletion:", friends)
name = input("\nEnter friend name to modify: ")
if name in friends:
    phone = input("Enter new phone number: ")
    friends[name] = phone
print("Updated dictionary:", friends)
name = input("\nEnter name to check: ")
if name in friends:
    print("Friend exists")
else:
    print("Friend not found")
print("\nSorted dictionary:")
for name in sorted(friends):
    print(name, ":", friends[name])
```

##### 11. Word Frequency in a Paragraph

```python
a = input("Enter paragraph: ")
words = a.split()
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1
print("Word Frequency:", freq)
```

##### 12. Find Common Keys in Two Dictionaries

```python
d1 = {'a':1, 'b':2, 'c':3}
d2 = {'b':5, 'c':7, 'd':8}
comm = []
for key in d1:
    if key in d2:
        comm.append(key)
print("Common keys:", comm)
```

##### 13. Invert Dictionary with Duplicate Value Input: d = {'a':1, 'b':2, 'c':1,’d’:3, 'e':2} Output: {1: ['a', 'c'], 2: ['b', 'e'], 3: ['d']}

```python
d = {'a':1, 'b':2, 'c':1, 'd':3, 'e':2}
result = {}
for key, value in d.items():
    if value not in result:
        result[value] = []
    result[value].append(key)
print(result)
```

##### 14. Using dictionary, print the first character that does not repeat. Input: "aabbccddeefg" Output: f

```python
s = "aabbccddeefg"
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
for ch in s:
    if freq[ch] == 1:
        print("First non repeating character:", ch)
        break
```

##### 16. Create a dictionary where keys are word lengths and values are lists of words. Input: ["cat","dog","elephant","rat","tiger"]  Output: {3:['cat','dog','rat'], 5:['tiger'],8:['elephant']}

```python
x = ["cat","dog","elephant","rat","tiger"]
res = {}
for w in x:
    l = len(w)
    if l not in res:
        res[l] = []
    res[l].append(w)
print(res)
```
