##### 1. Write a function that returns the    sum of digits    of a number using recursion. 

```python
def twosum(a,b):
    if a == 0:
        return b
    return twosum(a-1,b+1)
print(twosum(4,5))
```

##### 2. Write a function that checks whether a number is an    Armstrong number    (without using built in power function). 

```python
def chkams(n):
    x = str(n)
    d = len(x)
    total = 0
    temp = n
    while temp > 0:
        digit = temp % 10
        power = 1
        for _ in range(d):
            power *= digit
        total += power
        temp //= 10
    return total == n

chkams(153)
```

##### 3. Write a function that returns the    largest digit    in a number (without converting to string). 

```python
def lar(x):
    high = 0
    while x != 0:
        a = x % 10
        high = max(high, a)
        x //= 10
    return high

lar(125943)
```

##### 4. Write a function that takes a number and returns True if the number is palindrome and False otherwise (without using string conversion). 

```python
def chkpal(n):
    if n < 0:
        return False
    if n < 10:
        return True
    if n % 10 == 0 and n != 0:
        return False
    rev = 0
    while n > rev:
        rev = rev * 10 + n % 10
        n //= 10
    return n == rev or n == rev // 10

chkpal(1156511)
```

##### 5. Write a function that returns the    count of prime digits    in a number. 

```python
def primedgt(x):
    while(x != 0):
        a = x % 10
        if a in [2,3,5,7]:
            print(a, end=" ")
        x //= 10

primedgt(124578)
```

##### 6. Write a function that rotates digits of a number    k times to the right   . Example: rotate(12345,2) → 45123

```python
def rotk(n,k):
    a = str(n)
    if k % len(a) == 0:
        return n
    else:
        print(a[-k:]+a[:-k])
    
rotk(12345,2)
```

##### 7. Write a function that returns the factorial of a number using recursion without using loops.

```python
def fact(x):
    if x <= 1:
        return 1
    else:
        return x * fact(x-1)

print(fact(5))
```

##### 8. Write a function that checks whether a number is a Perfect Number   

```python
def perfect(n):
    if n < 1:
        return False
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s == n

perfect(6)
```

##### 9. Write a function that returns the    GCD of two numbers    using recursion.

```python
def gcd(a,b):
    if a == 0:
        return b
    return gcd(b%a, a)

gcd(12,18)
```

##### 10. Write a function that returns the    LCM of two numbers    using the GCD function. 

```python
def gcd(a,b):
    if a == 0:
        return b
    return gcd(b%a, a)
def lcm(a,b):
    return a*b // gcd(a,b)

lcm(12,16)
```

##### 13. Write a function that removes    duplicate elements    from a list without using `set()`. 

```python
def unique(l):
    res = []
    for i in l:
        if i not in res:
            res.append(i)
    return res
print(unique([1,3,3,4,5,5,6,1,7,8,2,9]))
```

##### 14. Write a function that returns the    frequency of each character    in a string. 

```python
def freq(s):
    d={}
    for i in set(s):
        d[i] = {s.count(i)}
    return d

freq("hiihello")
```

##### 16. Write a function that returns the    sum of factorial of each digit    of a number.

```python
def fact(x):
    if x <= 1:
        return 1
    else:
        return x * fact(x-1)
def factsum(n):
    s = 0
    while(n != 0):
        a = n % 10
        s += fact(a)
        n //= 10
    return s

factsum(25)
```
