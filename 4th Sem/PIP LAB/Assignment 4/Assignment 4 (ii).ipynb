##### 3. Write a Python program to sort a list of elements using the bubble sort algorithm.

```python
l = [1,9,1,4,5,9,7,10,9,8]
n = len(l)
for i in range(n):
    for j in range(0, n-i-1):
        if l[j] > l[j+1]:
            l[j], l[j+1] = l[j+1], l[j]
print(l)
```

##### 4. Write a Python program to find the common elements between two lists. 

```python
l1 = [1,9,1,44,5,9,17,10,9,8]
l2 = [10,9,11,44,57,19,17,10,9,81]
print(list(set(l1) & set(l2)), end=" ")
```

##### 5. Write a Python program to count the frequency of each element in a list. 

```python
l = [1,9,17,44,5,9,17,10,9,8]
l1 = list(set(l))
for i in l1:
    print("Frequency of ",i,"is",l.count(i))
```

##### 6. Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings

```python
l = ["HELLO", "HII", "OO", "TAT", "ISI"]
for ch in l:
    if(len(ch) >= 2  and ch[0] == ch[-1]):
        print(ch, end=" ")
```

##### 9. Write a Python program to generate and print a list of first and last 5 elements where the values are square of numbers between 1 and 30
