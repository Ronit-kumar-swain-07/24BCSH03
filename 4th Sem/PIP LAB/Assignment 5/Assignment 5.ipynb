##### 10. Take a tuple of mixed data types and find the square of each element.  

```python
t = (1, 'HII', 1.5, 'a', "Hello", '#', 10)
t1 = tuple(x**2 if isinstance(x, (int, float)) else x*2 for x in t)
print(t1)   
```

##### 13. Given a tuple of numbers, create another tuple such that it contains only the even numbers of the first tuple.

##### 14. Display the largest number of a tuple without using the built-in function max().The user need to input the values in tuple from the keyboard. 

```python
t = tuple(map(int,input("Enter tuple Values: ").split()))
x = t[0]
for i in t:
    if i > x:
        x = i
print("Largest value is : ",x)
```

##### 15. Replace the last value of the tuple with 100.Sample list: [(1, 2, 3), (4, 5, 6), (7, 8, 9)] Output: [(1, 2, 100), (4, 5, 100),    (7, 8, 100)]  

```python
t = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]
t1 = []
for i in t:
    t1.append(i[:-1] + (100,))
print(t1)
```

##### 16. Sort a tuple by its float element. Sample data: [('item1', '12.20'), ('item2','15.10'), ('item3', '24.5')] Output: [('item3', '24.5'), ('item2', '15.10'),  ('item1', '12.20')] 

```python
t = [('item1', '32.20'), ('item2', '15.10'), ('item3', '64.5')]
t = [[float(i[1]), i] for i in t]
t.sort(reverse = True)
print([i[1] for i in t])
```

##### 17. Count the number of elements in a tuple until an element is a list. 

```python
t = (11, 22,[i,3,67])
count = 0
for i in t:
    if type(i) == list:
        break
    count += 1
print(count)
```
