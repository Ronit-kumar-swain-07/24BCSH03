2. Rectangle Area
Create a class Rectangle with:
• length
• width

Add methods:
• area() → returns area
• perimeter() → returns perimeter

```python
class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2 * (self.length + self.width)

r1 = Rectangle(10,5)
a,p = r1.area(), r1.perimeter()
print("Area: ",a)
print("Perimeter: ",p)
```

4. Circle Class
Create a class Circle with attribute radius.
Add methods:
• area()
• circumference()
Use the value π = 3.14.

```python
class Circle:
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return 3.1415 * self.radius * self.radius
    def circumference(self):
        return 2 * 3.1415  *self.radius
    
c1 = Circle(5)
a,c = c1.area(), c1.circumference()
print("Area: ",a)
print("Circumference: ",c)
```

5. Bank Account System
Create a class BankAccount with:
• account_holder
• balance;
Methods:
• deposit(amount)
• withdraw(amount)
• display_balance()
Add a condition so withdrawal cannot exceed balance.

```python
class BankAccount:
    def __init__(self,account_holder,balance):
        self.account_holder = account_holder
        self.balance = balance
    def deposit(self,amount):
        self.balance += amount
    def withdraw(self,amount):
        if amount > self.balance:
            print("Withdraw Not Possible.")
        else:
            self.balance -= amount
    def display(self):
        print("Current Balance: ",self.balance)
b1 = BankAccount("Rahul",1000)
b1.deposit(2000)
b1.withdraw(500)
b1.display()
```

```python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
    def apply_discount(self, percent):
            discount = (percent / 100) * self.price
            self.price -= discount
b1 = Book("Python", "Rahul", 500)
print("Original Price:", b1.price)
b1.apply_discount(15)
print("Discounted Price:", b1.price)
```

8. Temperature Converter
Create a class Temperature with attribute celsius.
Methods:
• to_fahrenheit()
• to_kelvin()

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius
    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32
    def to_kelvin(self):
        return self.celsius + 273.15
t = Temperature(25)
print("Celsius:", t.celsius)
print("Fahrenheit:", t.to_fahrenheit())
print("Kelvin:", t.to_kelvin())
```

9. Library Management
Create a class Library with:
• book_list
Methods:
• add_book(book)
• remove_book(book)
• display_books()


```python
class Library:
    def __init__(self):
        self.book_list = []
    def add_book(self, book):
        self.book_list.append(book)
    def remove_book(self, book):
        if book in self.book_list:
            self.book_list.remove(book)
        else:
            print("Book not found")
    def display_books(self):
        if len(self.book_list) == 0:
            print("Library is empty")
        else:
            print("Available Books:")
            for b in self.book_list:
                print(b)
l = Library()
l.add_book("Python")
l.add_book("Java")
l.add_book("DBMS")
l.display_books()
l.remove_book("DBMS")
l.display_books()
```

10. Shopping Cart
Create a class Cart with:
• items (dictionary)
Methods:
• add_item(item, price)
• remove_item(item)
• total_price()

```python
class Cart:
    def __init__(self):
        self.items = {}
    def add_item(self, item, price):
        self.items[item] = price
    def remove_item(self, item):
        if item in self.items:
            del self.items[item]
        else:
            print("Item is not in cart")
    def total_price(self):
        total = sum(self.items.values())
        return total
c = Cart()
c.add_item("A", 100)
c.add_item("B", 400)
c.add_item("C", 500)
print("Total Price:", c.total_price())
c.remove_item("C")
print("Total Price after removal:", c.total_price())
```

11. Student Grade System
Create a class Student with:
• name
• marks (list)
Methods:
• average_marks()
• grade()
• high_grad()
Grades:
• A → 90+
• B → 75+
• C → 50+
• F → below 50


```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks        
    def average_marks(self):
        if len(self.marks) == 0:
            return 0
        return sum(self.marks) / len(self.marks)
    def grade(self):
        avg = self.average_marks()
        if avg >= 90:
            return 'A'
        elif avg >= 75:
            return 'B'
        elif avg >= 50:
            return 'C'
        else:
            return 'F'
    def high_mark(self):
        if len(self.marks) == 0:
            return None
        return max(self.marks)
    
s1 = Student("Raaj", [85, 90, 78, 92])
print("Name:", s1.name)
print("Marks:", s1.marks)
print("Average:", s1.average_marks())
print("Grade:", s1.grade())
print("Highest Marks:", s1.high_mark())
```

12. ATM Simulation
Create a class ATM with:
• balance
• pin
Methods:
• check_balance()
• deposit()
• withdraw()
• change_pin()

```python
class ATM:
    def __init__(self,balance,pin):
        self.balance = balance
        self.pin = pin
    def check_balance(self):
        print("Current Balance: ",self.balance)
    def deposit(self,amount):
        self.balance += amount
    def withdraw(self,amount):
        if self.balance > amount:
            self.balance -= amount
        else:
            print("ERROR!!!")
    def change_pin(self,x):
        self.pin = x
    def check_pin(self):
        print("Current Pin: ",self.pin)

a = ATM(1000,123789)
a.deposit(2000)
a.withdraw(500)
a.change_pin(741963)
a.check_balance()
a.check_pin()
```

13. Stack Data Structure Class with Display Method
Write a Python program to create a class representing a stack data structure. Include methods for
pushing, popping and displaying elements

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self,x):
        self.stack.append(x)
    def pop(self):
        if len(self.stack) == 0:
            print("Stack UnderFlow")
        else:
            self.stack.pop()
    def display(self):
        if len(self.stack) == 0:
            print("Stack Underflow")
        else:
            print(self.stack)

s1 = Stack()
s1.push(1)
s1.push(3)
s1.push(4)
s1.display()
s1.pop()
s1.display()
```

14. Person Class with Age Calculation
Create a Person class with attributes like name and date_of_birth . Implement a
method that calculates the person's current age based on the date of birth.

```python
from datetime import date
class Person:
    def __init__(self,name,dob):
        self.name = name
        self.dob = dob
    def calAge(self):
        td = date.today()
        return td.year - self.dob
    
p = Person("Rahul",1995)
print("Age:",p.calAge(),"years")
```

15. Complex numbers Addition
 Define a class Complex with attributes:
• real
• imaginary
Create methods to:
• input two complex numbers
• add them
• display the result

```python
class Complex:
    def __init__(self,r,i):
        self.r = r
        self.i = i
    def compadd(self,c2):
        R = self.r + c2.r
        I = self.i + c2.i
        return Complex(R,I)
    def display(self):
        print("Res: ",self.r,"+",self.i,"i")

c1 = Complex(2,5)
c2 = Complex(5,5)
res = c2.compadd(c1)
res.display()
```
